## Dynamic Programming

memoization 기법을 이용하여 이전에 계산한 값을 저장하여 다시 계산하지 않도록 함(피보나치)

## Divide and Conquer
문제를 나눌 수 없을 때까지 나누어서 다시 합병함. 

일반적으로 재귀함수로 구현하고 문제를 쪼갤때 부분 문제들은 서로 중복되지 않음(병합 정렬, 퀵 정렬 등)




```
# fibonacci
# using divide and conquer
def fibo(num):  
  if num <= 1:
    return num
  
  return fibo(num - 1) + fibo(num - 2)
```

```
# using DP
def fibo_dp(num):
  cache = [0 for index in range(num + 1)]
  cache[0] = 0
  cache[1] = 1

  for index in range(2, num + 1):
    cache[index] = cache[index - 1] + cache[index - 2] 
  
  return cache[num]
```
