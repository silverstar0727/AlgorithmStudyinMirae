### example1) 공유기설치문제
https://www.acmicpc.net/problem/2110

1. 범위: 제일 작은 폭 ~ 제일 큰폭(1 ~ 8)
2. 중앙값을 설정하고 이것보다 크면 설치 -> 설치가 안될 경우 그것보다 큰 값들을 버림(4를 설정, 1,8,, 설정 불가 4보다 큰 폭은 버림)
3. 범위 재설정(1~3)
4. 다시 중앙값설정(2 -> 1,4,8 = 됨, 1은 최대가 될 수 없으므로 버림)
5. 3이 되는가 봄 (3 -> 1, 4, 8 = 됨,  2는 최대가 될 수 없으므로 버림)
6. 성공

```python
n,c = map(int,input().split())

house = []
for _ in range(n):
    x = int(input())
    house.append(x)

house.sort()

# 최소 거리
start = house[1] - house[0]
# 최대 거리
end = house[-1] - house[0]

result = 0

while (start <= end):
    mid = (start+end)//2 # 해당 gap
    old = house[0]
    count = 1
    
    for i in range(1, len(house)):
        if house[i] >= old+mid: # gap 이상
            count+=1
            old = house[i]
    
    if count >=c:
        start = mid + 1
        result = mid
    else:
        end = mid - 1

print(result)
```
