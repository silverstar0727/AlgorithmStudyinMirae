### 트리(TREE)
* 트리: node와 branch를 이용하여, 사이클을 이루지 않도록 구성한 데이터 구조
* 이용 예: 이진 트리(binary tree) - 탐색 알고리즘 구현
* 용어
1.  node: 트리에서 데이터를 저장하는 기본 요소(다른 노드에 대한 연결 branch포함)
2. root bode: 트리 맨 위에 있는 노드
3. level: 최상위 노드 = level 0, 아래로 내려갈 수록 level + 1
4. parent node: 어떤 노드의 이전 레벨에 연결된 노드
5. child node: 어떤 노드의 다음 레벨에 연결된 노드
6. leaf node(terminal node): child node가 하나도 없는 노드
7. sibling (brother node): 동일한 parent node를 가진 노드
8. depth: 최대 level

### 이진 트리와 이진탐색 트리
* 이진 트리: 노드별 최대 branch가 2인 트리
* 이진 탐색 트리: 이진트리에 왼쪽 노드는 해당 노드보다 작은 값, 오른쪽 노드는 해당 노드보다 큰 값을 가진 트리

### 이진탐색 트리의 장점과 주요 용도
* 주요 용도: 데이터 검색
* 장점: 탐색 속도를 개선할 수 있음

#### 링크드 리스트로 이진트리를 구현하기

![image](https://user-images.githubusercontent.com/49096513/89526109-35b89b80-d822-11ea-946b-1e86a30a2698.png)


```
# 노드 클래스 만들기
# 각 노드는 자식노드의 왼쪽과 오른쪽을 가지고, value를 가짐
class Node: 
  def __init__(self, value):
    self.value = value
    self.left, self.right = None, None
```

1. 현재의 값이 더 큼

1.1 왼쪽이 있음 -> 왼쪽으로 이동

1.2 왼쪽이 비어있음 -> 왼쪽에 노드를 채움

2. 넣고자 하는 값이 더 크거나 같음

2.1 오른쪽이 있음 -> 오른쪽으로 이동

2.2 오른쪽이 비어있음 -> 오른쪽에 노드를 채움

```
# node management
# 노드를 채우는 class
class NodeMgmt:
  def __init__(self, head):
    self.head = head # 노드의 첫 값을 head로 설정

  def insert(self, value): # value는 집어 넣고자 하는 값
    self.current_node = self.head # current노드를 head로 설정

    while True:
      if value < self.current_node.value: # 현재의 값이 더 클 때
        if self.current_node.left != None: # 왼쪽의 노드가 존재할때
          self.current_node = self.current_node.left # 왼쪽으로 이동
        else: # 왼쪽의 노드가 존재하지 않을 때
          self.current_node.left = Node(value) # 넣고자 하는 값을 왼쪽으로 넣기
          break

      else: # 넣고자 하는 값이 값이 더 크거나 같을 때
        if self.current_node.right != None: # 오른쪽의 노드가 존재할 때
          self.current_node = self.current_node.right # 오른쪽으로 이동
        
        else: # 오른쪽의 노드가 존재하지 않을 때
          self.current_node.left = Node(value)
          break # 넣고자 하는 값을 오른쪽으로 넣기
```

```
head = Node(1)
binary_tree = NodeMgmt(head)
```

## 이진 탐색 트리 삭제
1. leaf node 삭제
* leaf node: child node가 없는 말단 node
* 삭제할 노드의 부모 노드가 삭제할 노드를 가리키지 않도록 함
2. child node가 하나인 노드의 삭제
* 삭제할 node의 부모노드가 삭제할 node의 자식노드를 가리키도록 함
3. 자식노드가 두개인 노드의 삭제
* 삭제할 노드의 오른쪽 자식 중 가장 작은 값을 삭제할 노드의 부모 노드가 가리키도록 한다
* 삭제할 노드의 왼쪽 자식중 가장 큰 값을 삭제할 노드의 부모노드가 가리키도록 한다.
