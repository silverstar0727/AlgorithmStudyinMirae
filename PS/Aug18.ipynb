```python
# 내풀이... 틀림 ㅠㅠ 런타임 에러
import numpy as np
n = int(input())
data = [list(map(int, list(input()))) for i in range(n)]
check_list = np.array([[False] * n] * n)
graph = {}

def make_graph(cnt, x, y):
    graph[cnt].append([x, y])
    check_list[y][x] = True
    
    # 오른쪽
    if x < n-1:
        if check_list[y][x+1] == False:
            if data[y][x+1] == 1:
                make_graph(cnt, x+1, y)
            else:
                check_list[y][x+1] = True
    # 왼쪽
    if x > 0:
        if check_list[y][x-1] == False:
            if data[y][x-1] == 1:
                make_graph(cnt, x-1, y)
            else:
                check_list[y][x-1] = True
    
    # 아래
    if y < n - 1:
        if check_list[y+1][x] == False:
            if data[y+1][x] == 1:
                make_graph(cnt, x , y+1)
            else:
                check_list[y+1][x] = True
    
    # 위
    if y > 0:
        if check_list[y-1][x] == False:
            if data[y-1][x] == 1:
                make_graph(cnt, x , y-1)
            else:
                check_list[y-1][x] = True
                
cnt = 0
for i in range(n):
    for j in range(n):
        if check_list[i][j] == True:
            continue
            
        if data[i][j] == 1: # 1일때
            cnt += 1
            graph[cnt] = []
            make_graph(cnt, j, i)
        else: # 0일때
            check_list[i][j] = True

print(len(graph))
l = []
for i in graph:
    l.append(len(graph[i]))
l.sort()
for i in l:
    print(i)
```

```python
# 솔루션
n = int(input())
s = []
dx = [1, -1, 0, 0]
dy = [0, 0, -1, 1]
cnt = []
for i in range(n):
    s.append(list(input()))
def bfs(i, j):
    queue = [[i, j]]
    s[i][j] = "0"
    count = 1
    while queue:
        a, b = queue[0][0], queue[0][1]
        del queue[0]
        for k in range(4):
            x = a + dx[k]
            y = b + dy[k]
            if 0 <= x < n and 0 <= y < n and s[x][y] == "1":
                s[x][y] = "0"
                queue.append([x, y])
                count += 1
    cnt.append(count)
for i in range(n):
    for j in range(n):
        if s[i][j] == "1":
            bfs(i, j)
cnt.sort()
print(len(cnt))
for i in cnt:
    print(i)
```
